3.Load the wavelet filter to the result of 2 and demonstrate the line chart

```python
import pywt   
from sklearn import preprocessing
```

```python
def maddest(d, axis=None):
    return np.mean(np.absolute(d - np.mean(d, axis)), axis)

def denoise_signal(x, wavelet='db4', level=1):
    coeff = pywt.wavedec(x, wavelet, mode="per")
    sigma = (1/0.6745) * maddest(coeff[-level])

    uthresh = sigma * np.sqrt(2*np.log(len(x)))
    coeff[1:] = (pywt.threshold(i, value=uthresh, mode='hard') for i in coeff[1:])

    return pywt.waverec(coeff, wavelet, mode='per')
```

```python
x_1=new_st['HOBBIES_1_050_CA_1_validation']
y_w1 = denoise_signal(x_1)
ts = pd.Series(y_w1,np.arange(200,502))
ts.plot(figsize=(20,5))
```

```python
dx=new_st.plot(x='Days',y='HOBBIES_1_050_CA_1_validation',legend='Original')
ts.plot(legend='AfterWavelet', title='Compare', ax=dx,figsize=(10,5))
```
